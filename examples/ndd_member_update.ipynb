# NeuroData Member Update Protocol
This notebook utilizes `pandarize` to parse people.bib file from neurodata.io and update the list

```python
from bibReader.frame import bReader
import pandas as pd
import bibReader
bibReader.__version__
```

```python
bib = bReader()
```

```python
bib.settings['convert_names'] = False
bib.settings['remove_empty_entries'] = False
```

```python
bib.load('https://raw.githubusercontent.com/neurodata/neurodata.io/deploy/content/bibs/people.bib')
bib.fit()
```

```python
filepath = 'dev/intermediate.xlsx'
ufilepath = 'dev/intermediate-updated.xlsx'
```

```python
bib.df.head()
```

```python
bib.df.to_excel(filepath, index=False, engine='xlsxwriter')
```

Make changes to the downloaded and converted file

```python
bib.df = pd.read_excel(ufilepath, dtype=object, na_filter=False)
```

```python
bib.df.head(5)
```

```python
bib.df.priority = bib.df.priority.astype(float)
bib.df.sort_values(['priority'], ascending=True, inplace=True)
```

```python
bib.transform(filename='people')
```
