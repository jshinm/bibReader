# NeuroData Member Update Protocol
This notebook utilizes `pandarize` to parse people.bib file from neurodata.io and update the list

```python
from pandarize.frame import Pandarizer
import pandas as pd
```

```python
pdr = Pandarizer()
pdr.load('https://raw.githubusercontent.com/neurodata/neurodata.io/deploy/content/bibs/people.bib')
pdr.fit()
```

```python
filepath = 'dev/intermediate.xlsx'
ufilepath = 'dev/intermediate-updated.xlsx'
```

```python
pdr.df.to_excel(filepath, index=False, engine='xlsxwriter')
```

```python
pdr.df = pd.read_excel(ufilepath, dtype='object')
```

```python
pdr.df.sort_values(['priority'], ascending=True, inplace=True)#.reset_index(drop=True)
```
