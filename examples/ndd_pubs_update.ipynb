# NeuroData Publication Update Protocol
This notebook utilizes `pandarize` to parse pubs.bib file from neurodata.io and update the list

```python
from bibReader.frame import bReader
import pandas as pd
import bibReader
bibReader.__version__
```

```python
bib = bReader()
bib.load('https://raw.githubusercontent.com/neurodata/neurodata.io/deploy/content/bibs/pubs.bib')
bib.fit()
```

```python
filepath = 'dev/intermediate.xlsx'
ufilepath = 'dev/intermediate-updated.xlsx'
```

```python
bib.df.to_excel(filepath, index=False, engine='xlsxwriter')
```

#### Check for duplicates

```python
bib.df[bib.df['alias'].duplicated()]
```

#### Make changes to the downloaded and converted file

```python
bib.df = pd.read_excel(ufilepath, dtype=object, na_filter=False)
```
